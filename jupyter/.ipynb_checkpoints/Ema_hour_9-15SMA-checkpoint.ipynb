```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Stock symbol
symbol = 'SBIN.NS'  # 'SBIN.BO' for BSE, 'SBIN.NS' for NSE
symbol = 'LTIM.NS'  # 'SBIN.BO' for BSE, 'SBIN.NS' for NSE

# Fetch historical data
stock_data = yf.download(symbol, start='2022-01-01', end='2023-01-01', interval='1h')
```

```python
stock_data
```

```python
df = stock_data
```

```python
df['EMA_9']=df['Close'].ewm(span=9, adjust=False).mean()
df['EMA_15']=df['Close'].ewm(span=15, adjust=False).mean()
df =df.reset_index()
```

```python
df
```

```python
cross_list = list()
stock_hold= False 
for index in range(df.shape[0]):
    close_value = str(df.loc[index,['Close']].values[0])
    if index<3:
        cross_list.append('None')
        # print(df.loc[22,['EMA_9']].values[0])
    elif (df.loc[index-2,['EMA_9']].values[0] > df.loc[index-2,['EMA_15']].values[0]) and (df.loc[index,['EMA_15']].values[0] > df.loc[index-2,['EMA_9']].values[0]):
        
        cross_list.append('BUY PE'+close_value)
        is_pe_buy = True
        # stock_hold= True
    elif df.loc[index-2,['EMA_9']].values[0] < df.loc[index-2,['EMA_15']].values[0] and df.loc[index,['EMA_15']].values[0] < df.loc[index-2,['EMA_9']].values[0]:
        cross_list.append('BUY_CE'+close_value)
        is_ce_buy = True
        # stock_hold= True
    else:
        # print(int(df.loc[index-2,['EMA_9']].values[0]) , int(df.loc[index-2,['EMA_15']].values[0]))
        cross_list.append("HOLD")
print(cross_list)
```

```python
import openpyxl
range = 120
high_limit = list()
high_cross = list()
high_sl = list()
high_fall = list()
low_limit = list()
low_cross = list()
low_sl=list()
both_sl_hit=list()
both_tgt_ach=list()
for index,row in df.iterrows():
    upper_limit = row.Open+range
    lower_limit = row.Open-range
    ##### Upper limit checking
    high_limit.append(upper_limit)
    if row.High>=upper_limit:
        high_cross.append(row.High-upper_limit)
        high_sl.append("SL Hit")
        high_fall.append(row.Open+range-row.Close)
        high_sl_hit = True
    else:
        high_cross.append(0)
        high_sl.append("None")
        high_fall.append("None")
        high_sl_hit = False
    ##### Lower limit checking    
    low_limit.append(lower_limit)
    if row.Low<=lower_limit:
        low_cross.append(lower_limit - row.Low)
        low_sl.append("SL Hit")
        low_sl_hit = True
    else:
        low_cross.append(0)
        low_sl.append("None")
        low_sl_hit = False
    if high_sl_hit == True and low_sl_hit == True:
        both_sl_hit.append("SL")
    else:
        both_sl_hit.append("None")
    if high_sl_hit == False and low_sl_hit == False:
        both_tgt_ach.append("TGT")
    else:
        both_tgt_ach.append("None")
    
        

df['Low_Limit']=low_limit
df['High_Limit']=high_limit
df['High_Cross']=high_cross
df['Low_Cross']=low_cross
df['High_SL']=high_sl
df['Low_SL']=low_sl
df['Both_SL_Hit'] = both_sl_hit
df['Both_TGT_Ach'] = both_tgt_ach
df['High_Fall']=high_fall

display(df)
no_of_days = len(both_tgt_ach)
no_of_tgt_ach = len([x for x in both_tgt_ach if x=='TGT'])
print("No of time TGT achieved within the Range of {0} point :: {1} days out of {2} days".format(range,no_of_tgt_ach,no_of_days))
win_per = round((no_of_tgt_ach/no_of_days)*100,2)
print("Win percentage ::{0}%".format(win_per))
df.to_excel("Sorting_Heding_200D_120P.xlsx",engine='openpyxl')
```
